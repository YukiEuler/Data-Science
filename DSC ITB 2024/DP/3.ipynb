```python
df = pd.read_csv('data2.csv')
df['new_target'] = df['Cholesterol Total (mg/dL)'].apply(
    lambda x: 1 if x == 187 else 0)
df.set_index('Responden', inplace=True)
df.drop(df[df['new_target'] == 1].index, inplace=True)
df.drop('new_target', axis=1, inplace=True)
df.head()
```

```python
# encoding Jenis Kelamin if M = 1 and F = 0

df["Jenis Kelamin"] = df["Jenis Kelamin"].apply(lambda x: 1 if x == 'M' else 0)

# drop Tempat lahir
df.drop('Tempat lahir', axis=1, inplace=True)
df.head()
```

```python
# Separate target variable and features
y = df['Cholesterol Total (mg/dL)']
X = df.drop('Cholesterol Total (mg/dL)', axis=1)

# Define the number of folds for cross-validation
num_folds = 50

# Initialize KFold with stratified sampling
kf = KFold(n_splits=num_folds, shuffle=True, random_state=42)

# Initialize models
models = {
    'XGBoost': XGBRegressor(),
    'Linear Regression': LinearRegression(),
    'Ridge Regression': Ridge(),
    'Lasso Regression': Lasso(),
    'Random Forest': RandomForestRegressor()
}

# Function to perform k-fold cross-validation


def perform_cross_validation(model, X, y, kf):
    # Initialize list to store RMSE scores for each fold
    rmse_scores = []

    # Perform k-fold cross-validation
    for train_index, test_index in kf.split(X, y):
        X_train, X_test = X.iloc[train_index], X.iloc[test_index]
        y_train, y_test = y.iloc[train_index], y.iloc[test_index]

        # Train the model
        model.fit(X_train, y_train)

        # Make predictions on the test set
        y_pred = model.predict(X_test)

        # Calculate RMSE
        rmse = np.sqrt(mean_squared_error(y_test, y_pred))

        # Append RMSE to the list of scores
        rmse_scores.append(rmse)

    return rmse_scores


# Perform cross-validation for each model
results = {}
for model_name, model in models.items():
    rmse_scores = perform_cross_validation(model, X, y, kf)
    average_rmse = np.mean(rmse_scores)
    results[model_name] = {'RMSE Scores': rmse_scores,
                           'Average RMSE': average_rmse}

# Print the results for each model
for model_name, result in results.items():
    print(model_name + ":")
    print("RMSE Scores:", result['RMSE Scores'])
    print("Average RMSE:", result['Average RMSE'])
    print()
```
